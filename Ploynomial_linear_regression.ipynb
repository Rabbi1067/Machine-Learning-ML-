```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#df=pd.read_csv('housing_data.csv')
```

Quadratic equation used- y= 0.5x^2 + 1.5x + 2 + outliers

```python
x=6*np.random.rand(100,1)-3
y=0.5*x**2 + 1.5*x + 2 + np.random.randn(100,1)
plt.plot(x,y,'b.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Implement simple linear regression

```python
from sklearn.linear_model import LinearRegression
model_1=LinearRegression()
```

```python
model_1.fit(x_train,y_train)
```

```python
x_train
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,model_1.predict(x_test))
print(score)
```

Lets visualize this model

```python
plt.plot(x_train,model_1.predict(x_train),'r-')
plt.scatter(x_train,y_train,color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

lets apply polynomial transformation

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly= PolynomialFeatures(degree=2,include_bias=True)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.fit_transform(x_test)
```

```python
x_train_poly
```

```python
x_test_poly
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
model_1=LinearRegression()
model_1.fit(x_train_poly,y_train)
score = r2_score(y_test,model_1.predict(x_test_poly))
print(score)
```

```python
print(model_1.coef_)
print(model_1.intercept_)
```

```python
plt.plot(x_train,model_1.predict(x_train_poly),'r-')
plt.scatter(x_train,y_train,color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Prediction of new data set

```python
x_new=np.linspace(-3,3,200).reshape(200,1)
x_new_poly=poly.fit_transform(x_new)
x_new_poly

```

```python
y_new=model_1.predict(x_new_poly)
plt.plot(x_new,y_new,'r-',linewidth=2,label='new predictions')
plt.plot(x_train,y_train,color='b',label='Traning points')
plt.plot(x_test,y_test,color='g',label='Testing points')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
#
```
