```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

Create a dataset

```python
X,y=make_classification(n_samples=1000, n_features=10,n_classes=2, n_informative=3, n_redundant=0, random_state=0)
```

```python

pd.DataFrame(X)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

Model train

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(x_train,y_train)
```

Prediction for test data

```python
y_pred=logistic.predict(x_test)
print(y_pred)
```

```python
logistic.predict_proba(x_test)
```

Perfromance matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
print(score)  # final output
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

**Hyperparameter Tuning And Cross Validation**

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear','sag','saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

GridsearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
cv=RepeatedStratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,cv=5,scoring='accuracy',n_jobs=-1)
#grid.fit(x_train,y_train)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
