Datatypes in csv

```python
data=('col1,col2,col3\n'
      '1,2,3\n'
      '4,5,6\n'
      '7,8\n')
```

```python
df=pd.read_csv(StringIO(data))
```

```python
df.info()
```

```python
df=pd.read_csv(StringIO(data),dtype='object')
```

```python
df.info()
```

```python
df=pd.read_csv(StringIO(data),dtype={'col1':int,'col2':float,'col3':float})
```

```python
df.info()
```
