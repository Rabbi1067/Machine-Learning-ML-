Scatter plot

```python
sns.scatterplot(data=pen,x='flipper_length_mm',y='body_mass_g',hue='sex')
```

```python
sns.set_style('dark') #background style
sns.scatterplot(data=pen,x='flipper_length_mm',y='body_mass_g',hue='sex')
```

```python
sns.set_style('ticks')
sns.scatterplot(data=pen,x='flipper_length_mm',y='body_mass_g',hue='sex')
#sns.despine(left=True)
```

**Context**

```python
sns.set_style('ticks')
sns.set_context("poster") #talk,paper,notebook etc parametter
sns.scatterplot(data=pen,x='flipper_length_mm',y='body_mass_g',hue='sex')
sns.despine(left=True)
```

**Palette**

```python
sns.set_style('ticks')
sns.set_context("notebook") #talk,paper,notebook etc parametter
sns.scatterplot(data=pen,x='flipper_length_mm',y='body_mass_g',hue='island',palette="Set1") #set1,set2,set3,dark etc palette
sns.despine()
```

```python
sns.scatterplot(data=pen,x='species',y='body_mass_g',hue='island',style='sex',alpha=0.5)
```

```python
sns.stripplot(data=pen,x='species',y='body_mass_g',hue='island',dodge=True,jitter=True)
```

Swarm plot

```python
sns.swarmplot(data=pen,x='species',y='body_mass_g',hue='island')
```

Histogram

```python
sns.set_context("notebook")
sns.histplot(data=pen,x='body_mass_g',hue='sex',multiple='stack')
```

Regresssion Plot

```python
sns.regplot(data=pen,x='body_mass_g',y='flipper_length_mm',color="green")
```

Line plot

```python
sns.set_style("whitegrid")
sns.lineplot(data=pen,x='body_mass_g',y='flipper_length_mm',hue='island',style='sex')
```

Joint plot

```python
sns.jointplot(data=pen,x="body_mass_g",y="flipper_length_mm")
```

```python
sns.jointplot(data=pen,x="body_mass_g",y="flipper_length_mm",hue='sex') #,kind='kde'
```

Bar plot

```python
sns.barplot(data=pen,x="species",y="flipper_length_mm",hue='sex')
```

```python
sns.barplot(data=pen,x="species",y="flipper_length_mm",hue='sex',palette=['pink','skyblue'],estimator=np.median)
```

Count plot

```python
sns.countplot(data=pen,x='species',hue='island')
```

Box plot

```python
sns.boxplot(data=pen,x='species',y='flipper_length_mm',hue='sex')
```

Violin plot

```python
sns.violinplot(data=pen,x='species',y='flipper_length_mm',hue='sex')
```

```python
sns.violinplot(data=pen,x='species',y='body_mass_g')
sns.swarmplot(data=pen,x='species',y='body_mass_g',color='black',size=3)
```

KDE plot

```python
sns.kdeplot(data=pen,x='flipper_length_mm',hue='sex')
```
