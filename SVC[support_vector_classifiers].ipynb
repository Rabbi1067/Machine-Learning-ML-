```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # for data visualization
import seaborn as sns # for statistical data visualization
%matplotlib inline
```

Lets create synthetic data points

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC (kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

Prediction

```python
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

If data overlapping

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC (kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
svc.coef_
```

```python
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred)) # accuracy decrease
print(confusion_matrix(y_test,y_pred))
```

So we use difference type of kernel

RBF kernel

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(x_train,y_train)
```

```python
y_pred1=rbf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred1)) # accuracy Increase
print(confusion_matrix(y_test,y_pred1))
```

Polynomial kernel

```python
polynomial=SVC(kernel='poly')
```

```python
polynomial.fit(x_train,y_train)
```

```python
y_pred2=polynomial.predict(x_test)
```

```python
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

Sigmoid

```python
sigmoid_svc = SVC(kernel='sigmoid')
sigmoid_svc.fit(x_train, y_train)
y_pred_sigmoid = sigmoid_svc.predict(x_test)
print(classification_report(y_test, y_pred_sigmoid))
print(confusion_matrix(y_test, y_pred_sigmoid))
```

Grid

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for GridSearchCV
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf']
}

# Create an SVC object
grid_svc = SVC()

# Initialize GridSearchCV
grid = GridSearchCV(grid_svc, param_grid, refit=True,cv=5, verbose=2)

# Fit the GridSearchCV object to the training data
grid.fit(x_train, y_train)
```

```python
# Print the best parameters and best score
print("Best Parameters found:", grid.best_params_)
print("Best Score found:", grid.best_score_)

# Make predictions using the best estimator
grid_predictions = grid.predict(x_test)

# Evaluate the best estimator
print(classification_report(y_test, grid_predictions))
print(confusion_matrix(y_test, grid_predictions))
```
