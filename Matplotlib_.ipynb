```python
import matplotlib.pyplot as plt
```

Chart

```python
x=[1,2,3,4,5]
y=[10,20,25,35,45]
```

line chart

```python
plt.plot(x,y)
plt.show()
```

### Customize ***chart***

```python

plt.figure(figsize=(10,5)) #figure size
plt.plot(x,y,color="green",marker='o',linestyle='dashed',linewidth=2,markersize=12)
plt.title("CSE is Fun")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

Advanced line chart- multiple lines & Legends

```python
x=[1,2,3,4,5]
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]
```

```python
plt.plot(x,y1,label='sales 2024') #ploting y1 data
plt.plot(x,y2,label='sales 2025') # ploting y2 data
plt.title("CSE is Fun")
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()#show legends
plt.show()
```

Bar chart

```python
plt.bar(x,y)
plt.title("CSE is fun")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

Histogram[distribution analysis]

```python
#data
import random
data= [random.randint(1,10) for _ in range(500)]
```

```python

plt.hist(data,bins=20) #bins means koi part a data dekhabe
plt.title("Histogram Example")
plt.show()
```

Pie chart[used to show part to whole relationships/data]

```python
catagoris=['A','B','C','D','E']
sales=[10,20,25,35,45]
```

```python
plt.pie(sales,labels=catagoris,autopct='%1.1f%%',startangle=90)
plt.title("CSE is Fun")
plt.show()
```

Scatter plt[used to find relation between varbles]

```python
plt.scatter(y1,y2)
plt.title("CSE is Fun")
plt.show()
```

subplots[used to show multiple charts in one figure]

```python
#data-1 bar chart
catagoris=['Sat','Sun','Mon','Tue','Web']
sales=[10,20,25,35,45]

#data-2 sactter chart
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]
```

```python
plt.figure(figsize=(6,4))
#first plot bar chart
plt.subplot(1,2,1) #row,column,position
plt.bar(catagoris,sales)
plt.title("Weekly Sales")
plt.xlabel("week days")
plt.ylabel("sales")


#second plot scatter chart
plt.subplot(1,2,2) #row,column,position
plt.scatter(y1,y2)
plt.title("User  example")
plt.xlabel("user1 ")
plt.ylabel("user2")

plt.show()
```
